```python
data  =pd.DataFrame({'Ratio':[len(df[df['Gender'] == 'M']), len(df[df['Gender'] == 'F'])]},
                      index = ['Male', 'Female'])

data.plot.pie(y = 'Ratio', figsize = (6,6) , autopct = "%.1f%%")
```

```python
df.groupby('Gender').size().plot(kind = 'bar', figsize = (8,8))
```

```python
data= df.groupby('Age')["Product_ID"].nunique().reset_index()
data.columns= ["age","Products"]

# Define the custom order for 'Age' categories
age_order = ['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+']
# Convert 'age' column to a categorical type with the specified order
data['age'] = pd.Categorical(data['age'], categories=age_order, ordered=True)
# Sort the DataFrame by the ordered 'age' column to ensure correct plotting order
data = data.sort_values('age')
```

```python
data.plot.bar(x='age', figsize=(8,6))
plt.show()
```
